```python
import cv2
import matplotlib.pyplot as plt
import selectivesearch.selectivesearchctivesearch as ss
```

```python
img = cv2.imread("sample.jpg")
img_lbl, regions = ss.selective_search(img, scale = 12000, sigma = 0.8, min_size = 200)
for r in regions:
    x, y, h, w = r['rect']
    cv2.rectangle(img, (x, y), (x + h, y + w), (0, 255, 0), 2)
plt.imshow(img)
plt.show()
```
